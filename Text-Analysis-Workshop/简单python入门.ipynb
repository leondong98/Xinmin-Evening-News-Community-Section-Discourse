Âú® Python ‰∏≠, Á¥¢ÂºïÊòØËÆøÈóÆÊï∞ÊçÆÂÆπÂô®‰∏≠Âçï‰∏™ÂÖÉÁ¥†ÁöÑÊñπÂºè, ÈÄÇÁî®‰∫éÂàóË°®(list)„ÄÅÂÖÉÁªÑ(tuple)„ÄÅÂ≠óÁ¨¶‰∏≤(string)ÂíåÂ≠óÂÖ∏(dictionary)Á≠âÊï∞ÊçÆÁªìÊûÑ„ÄÇ

Python ‰ΩøÁî®‰ªé 0 ÂºÄÂßãÁöÑÁ¥¢Âºï(zero-based indexing), ‰πüÂ∞±ÊòØËØ¥:

  -	Á¨¨‰∏Ä‰∏™ÂÖÉÁ¥†ÁöÑ‰ΩçÁΩÆÊòØ 0, Á¨¨‰∫å‰∏™ÊòØ 1, ‰æùÊ≠§Á±ªÊé®„ÄÇ

  - Â¶ÇÊûúËÆøÈóÆ‰∏Ä‰∏™‰∏çÂ≠òÂú®ÁöÑÁ¥¢Âºï‰ΩçÁΩÆÔºåÂ∞Ü‰ºöÂºïÂèëÁ¥¢ÂºïÈîôËØØ (IndexError)„ÄÇ

‰∏éÊ≠§‰∏çÂêå, Â≠óÂÖ∏(dictionary)‰ΩøÁî®ÁöÑÊòØÈîÆÁ¥¢Âºï(key-based indexing), ËÄå‰∏çÊòØ‰ΩçÁΩÆÁ¥¢Âºï„ÄÇ
‰Ω†ÈÄöËøáÈîÆÊù•ËÆøÈóÆÂØπÂ∫îÁöÑÂÄºÔºåËÄå‰∏çÊòØÈÄöËøáÊï∞Â≠ó‰ΩçÁΩÆ„ÄÇ

| Êï∞ÊçÆÁ±ªÂûã       | Á¥¢ÂºïÁ±ªÂûã        | ÊòØÂê¶ÊúâÈ°∫Â∫è | ÊîØÊåÅË¥üÁ¥¢Âºï | Á¥¢ÂºïÊòØÂê¶‰∏∫Êï¥Êï∞ | Á§∫‰æã                          |

### **Python ÂàáÁâá (Slicing)**

**ÂàáÁâáÊòØ‰ªéÂ∫èÂàó‰∏≠ËÆøÈóÆ‰∏ÄÈÉ®ÂàÜÂÖÉÁ¥†ÁöÑÊñπÂºè** ÈÄÇÁî®‰∫éÂàóË°®(list)„ÄÅÂÖÉÁªÑ(tuple)ÂíåÂ≠óÁ¨¶‰∏≤(string)Á≠âÂ∫èÂàóÁ±ªÂûã„ÄÇ

Âü∫Êú¨ËØ≠Ê≥ïÊòØ: `sequence[start:stop]`

 - ËøîÂõû‰ªé start Á¥¢ÂºïÂºÄÂßã„ÄÅÂà∞ stop Á¥¢Âºï‰πãÂâç(‰∏çÂåÖÊã¨ stop)ÁöÑ‰∏Ä‰∏™Êñ∞Â≠êÂ∫èÂàó„ÄÇ
 - Ê≥®ÊÑèÔºöÂàáÁâá‰∏≠ÁöÑ stop ÊòØ‚ÄúÂ∑¶Èó≠Âè≥ÂºÄ‚ÄùÁöÑÔºåÂç≥ÂåÖÊã¨Ëµ∑ÂßãÔºå‰∏çÂåÖÊã¨ÁªìÊùü„ÄÇ

‰Ω†ËøòÂèØ‰ª•Âä†ÂÖ•Ê≠•ÈïøÂèÇÊï∞, ËØ≠Ê≥ïÊòØ: `sequence[start:stop:step]`

- step Ë°®Á§∫ÊØèÊ¨°ÁßªÂä®Â§öÂ∞ë‰∏™Á¥¢ÂºïÂçï‰ΩçÔºåÈªòËÆ§ÊòØ 1„ÄÇ

````python
#@title Slicing Example
my_list = [0, 1, 2, 3, 4, 5]

print(my_list[1:4])      # ËæìÂá∫ [1, 2, 3]
print(my_list[:3])       # ËæìÂá∫ [0, 1, 2]
print(my_list[::2])      # ËæìÂá∫ [0, 2, 4]
print(my_list[::-1])     # ËæìÂá∫ [5, 4, 3, 2, 1, 0]ÔºåÂèçËΩ¨ÂàóË°®
````

ÁªßÊâøÊúâÂä©‰∫é‰øùÊåÅ‰ª£Á†ÅÁöÑÁªÑÁªáÊÄßÔºåÂáèÂ∞ëÈáçÂ§çÔºåÂπ∂ÂÆûÁé∞ÁÅµÊ¥ª„ÄÅÂèØÊâ©Â±ïÁöÑÁ®ãÂ∫èËÆæËÆ°„ÄÇ


````python
#@title Example Inheretance

class GraduateStudent(Student):
    def __init__(self, name, subject, grade, has_thesis):
        # Use super() to call the parent class's __init__ method
        super().__init__(name, subject, grade)
        self.has_thesis = has_thesis

    def is_doctor(self):
        # Graduate students need at least 70 to pass
        return self.grade >= 70 and self.has_thesis

anna = Student("Anna", "Python", 67)
bob = GraduateStudent("Bob", "AI Ethics", 72, True)
chris = GraduateStudent("Chris", "AI Ethics", 68, True)

print(anna.has_passed())   # True (Student needs 51)
print(bob.has_passed())    # True (Student needs 51)
print(chris.has_passed())  # True (Student needs 51)

print(bob.is_doctor())    # True (Graduate needs 70 + thesis)
print(chris.is_doctor())  # False (68 is not enough)

````

### **Êï∞ÁªÑÁöÑÁ¥¢Âºï‰∏éÂàáÁâáÔºàIndexing and Slicing ArraysÔºâ**

````python
#@title Example Array Indexing

arr_2d = np.arange(27).reshape((9,3))  #ÂàõÂª∫‰∏Ä‰∏™Êï∞Â≠ó‰ªé0Âà∞27Ôºå9Ë°å‰∏âÂàóÁöÑÁü©Èòµ
arr_2d

arr_2d[0, 0] # get first row first column value
arr_2d[0, -1] # last column of first row
arr_2d[1, ] # all of second row

arr_3d = np.arange(27).reshape((3,3,3))
arr_3d

arr_3d[1] # get second element of first dimension (matrix)
arr_3d[1, 1] # get second matrix, second row
arr_3d[:, -1, -1] # get second matrix, second row, third column value
````

### **Êï∞ÁªÑÁöÑÂàáÁâáÔºàSlicing ArraysÔºâ**
Ë¶ÅËé∑ÂèñÂ§ö‰∏™ÂÖÉÁ¥†ÊàñÂ≠êÊï∞ÁªÑÔºåÊàë‰ª¨‰ªçÁÑ∂‰ΩøÁî®ÊñπÊã¨Âè∑Ôºàsquare braketsÔºâÂíåÂàáÁâáËØ≠Ê≥ïÔºö


```
array[start:stop:step]
```

- `start` Ëµ∑ÂßãÁ¥¢ÂºïÔºàÈªòËÆ§ÂÄº‰∏∫ 0Ôºâ
- `top` ÁªàÊ≠¢Á¥¢ÂºïÔºàÈªòËÆ§ÊòØËØ•Áª¥Â∫¶ÁöÑÈïøÂ∫¶Ôºå‰∏çÂåÖÂê´ËØ•Á¥¢ÂºïÔºâ
- `step` Ê≠•ÈïøÔºàÈªòËÆ§ÊòØ 1Ôºâ

Â¶ÇÊûúÂ∞Ü `step` ËÆæÁΩÆ‰∏∫Ë¥üÂÄºÔºå`start` Âíå `stop` ÁöÑÈªòËÆ§ÂÄº‰ºöË¢´‰∫§Êç¢ÔºåËøôÊòØ‰∏ÄÁßç**ÂèçËΩ¨Êï∞ÁªÑ**ÁöÑÂ∏∏ËßÅÊäÄÂ∑ß„ÄÇ


> NumPy ÁöÑÂàáÁâáÊìç‰Ωú‰ºöÂàõÂª∫Êï∞ÁªÑÁöÑ‰∏Ä‰∏™ËßÜÂõæÔºàviewÔºâÔºåËÄå‰∏çÊòØÂâØÊú¨ÔºàcopyÔºâ„ÄÇËøôÊÑèÂë≥ÁùÄÂØπÂàáÁâáÂêéÁöÑÊï∞ÁªÑËøõË°åÊõ¥ÊîπÔºå‰ºöÂΩ±ÂìçÂéüÊï∞ÁªÑ‰∏≠ÁöÑÂØπÂ∫îÈÉ®ÂàÜ„ÄÇ


| Êìç‰Ωú            | Âê´‰πâ                         | Á§∫‰æã             | ËØ¥Êòé/Â§áÊ≥®                     |
|-----------------|------------------------------|------------------|------------------------------|
| `array[i, j]`   | ËÆøÈóÆÁ¨¨ i Ë°åÁ¨¨ j ÂàóÁöÑÂÖÉÁ¥†       | `a[1, 2]`        | Áî®‰∫é‰∫åÁª¥ÊàñÊõ¥È´òÁª¥Êï∞ÁªÑÁ¥¢Âºï     |
| `::2`           | Ê≠•Èïø‰∏∫ 2ÔºåÈöî‰∏Ä‰∏™Âèñ‰∏Ä‰∏™ÂÖÉÁ¥†     | `a[::2]`         | Áî®‰∫éÂàáÁâáÔºåËøîÂõûÂÅ∂Êï∞Á¥¢ÂºïÁöÑÂÖÉÁ¥† |
| `[::-1]`        | ÂÄíÂ∫è                           | `a[::-1]`        | Âø´ÈÄüÂèçËΩ¨Êï∞ÁªÑÈ°∫Â∫è             |
| `.copy()`       | ÊòæÂºèÂ§çÂà∂Êï∞ÁªÑ                   | `b = a.copy()`   | Èò≤Ê≠¢‰øÆÊîπËßÜÂõæÂΩ±ÂìçÂéüÊï∞ÁªÑ       |
| `.reshape()`    | ÊîπÂèòÊï∞ÁªÑÂΩ¢Áä∂Ôºå‰∏çÊîπÂèòÊï∞ÊçÆÂÜÖÂÆπ   | `a.reshape(2, 3)`| Êñ∞ÂΩ¢Áä∂ÂøÖÈ°ª‰∏éÂÖÉÁ¥†Êï∞ÈáèÂåπÈÖç     |

````python
#@title Example Slicing Arrays

arr_2d = np.arange(27).reshape((3, 9))     # ÂàõÂª∫‰∏Ä‰∏™‰ªé 0 Âà∞ 26 ÁöÑ‰∏ÄÁª¥Êï∞ÁªÑÔºåÈáçÂ°ëÊàê 3 Ë°å 9 ÂàóÁöÑ‰∫åÁª¥Êï∞ÁªÑ
arr_2d

arr_2d[1, :4] # Á¨¨ 2 Ë°åÔºàÁ¥¢Âºï‰∏∫ 1ÔºâÔºåÂâç 4 ‰∏™ÂÖÉÁ¥†
arr_2d[1, ::2] # Á¨¨ 2 Ë°åÔºåÊØèÈöî‰∏Ä‰∏™Âèñ‰∏Ä‰∏™ÔºàÊ≠•Èïø‰∏∫ 2Ôºâ
arr_2d[::2, ::2] # ÊØèÈöî‰∏ÄË°å & ÊØèÈöî‰∏ÄÂàóÂèñ‰∏Ä‰∏™ÔºåËµ∑ÁÇπÈÉΩÊòØÁ¨¨ 0 Ë°å/Âàó

arr_2d[::-1, ] # ÊâÄÊúâË°åÂèçËΩ¨Ôºà‰∏ä‰∏ãÈ¢†ÂÄíÔºâ

# view vs copy
second_row = arr_2d[1]  # ËøôÂπ∂‰∏çÊòØÂ§çÂà∂‰∫ÜÁ¨¨ 2 Ë°åÔºåËÄåÊòØÂàõÂª∫‰∫Ü ‰∏Ä‰∏™ÂØπÂéüÊï∞ÁªÑÁöÑ‚ÄúËßÜÂõæ‚ÄùÔºàviewÔºâ
second_row

second_row[:2] = 999 # Ê≥®ÊÑèÔºÅÊääËßÜÂõæ‰∏≠ÁöÑÂâç‰∏§‰∏™ÂÄºÊîπÊàê‰∫Ü 999ÔºåÂéüÊï∞ÁªÑ‰πü‰ºöË¢´‰øÆÊîπÔºÅ
second_row
arr_2d

#‚ö†Ô∏èÊ≥®ÊÑè‚ö†Ô∏è NumPy ÈªòËÆ§Áî®‚ÄúËßÜÂõæ‚ÄùËÄå‰∏çÊòØ‚ÄúÂ§çÂà∂‚ÄùÔºåËøôÂèØ‰ª•ËäÇÁúÅÂÜÖÂ≠òÔºå‰ΩÜÂ¶ÇÊûú‰Ω†Ë¶ÅÈÅøÂÖçÂΩ±ÂìçÂéüÊï∞ÊçÆÔºåËøòÊòØÈúÄ‰ΩøÁî® .copy()


# ÂàõÂª∫‰∏âÁª¥Êï∞ÁªÑ
arr_3d = np.arange(27).reshape((3,3,3)) #ÂàõÂª∫‰∏Ä‰∏™ÂΩ¢Áä∂‰∏∫ (3,3,3) ÁöÑ‰∏âÁª¥Êï∞ÁªÑÔºà3 ‰∏™ 3x3 ÁöÑÂ∞èÁü©ÈòµÔºâ
arr_3d

arr_3d[1, 1, ::2] # ÂèñÁ¨¨ 2 ‰∏™‚ÄúÁü©Èòµ‚ÄùÔºàÁ¥¢Âºï‰∏∫ 1ÔºâÂÆÉÁöÑÁ¨¨ 2 Ë°åÔºàÁ¥¢Âºï‰∏∫ 1ÔºâÔºåËøô‰∏ÄË°å‰∏≠ÔºåÊØèÈöî‰∏ÄÂàóÂèñ‰∏Ä‰∏™ÂÖÉÁ¥†ÔºàÊ≠•Èïø‰∏∫ 2Ôºâ
````

### **È´òÁ∫ßÁ¥¢ÂºïÔºàAdvanced IndexingÔºâ**

‚ÄúÈ´òÁ∫ßÁ¥¢Âºï‚ÄùÔºà‰πüÁß∞‰∏∫‚ÄúËä±ÂºèÁ¥¢Âºï‚Äù fancy indexingÔºâÊåáÁöÑÊòØ‰ΩøÁî®Êï∞ÁªÑÂØπÂè¶‰∏Ä‰∏™Êï∞ÁªÑËøõË°åÁ¥¢ÂºïÁöÑÊâÄÊúâÊÉÖÂÜµ„ÄÇ**È´òÁ∫ßÁ¥¢ÂºïÊÄªÊòØËøîÂõû‰∏Ä‰∏™ÂâØÊú¨ÔºàcopyÔºâÔºåËÄå‰∏çÊòØÂéüÂßãÊï∞ÊçÆÁöÑËßÜÂõæÔºàviewÔºâ„ÄÇ**

È´òÁ∫ßÁ¥¢Âºï‰Ωø‰Ω†ËÉΩÂ§ü**Âø´ÈÄüÂú∞ËÆøÈóÆÂíå‰øÆÊîπÊï∞ÁªÑ‰∏≠Êüê‰∫õÂÄºÁöÑÂ≠êÈõÜ**ÔºåÁâπÂà´ÈÄÇÁî®‰∫éÊåâÊù°‰ª∂ÊàñÁâπÂÆöÁ¥¢Âºï‰ΩçÁΩÆÊèêÂèñÂ§ö‰∏™ÂÖÉÁ¥†„ÄÇ

````python
#@title Example Fancy Indexing
arr_2d = np.arange(15).reshape(5,3)
arr_2d

idx = [1,2]

arr_2d[idx] # select indexed rows
arr_2d[:, idx] # select indexed columns

rows = [0, 1]
cols = [1, 2]

arr_2d[rows, cols]
````

**NumPy ‰∏≠ÁöÑÂ∏ÉÂ∞îÁ¥¢ÂºïÊìç‰ΩúÁ¨¶ÔºàBoolean Indexing OperatorsÔºâ**

| Êìç‰ΩúÁ¨¶  | Êìç‰ΩúÁ±ªÂûã        | Á§∫‰æã                          | ÊèèËø∞ËØ¥Êòé                         |
|---------|-----------------|-------------------------------|----------------------------------|
| `&`     | Êåâ‰Ωç‰∏éÔºàANDÔºâ   | `(arr > 5) & (arr < 10)`      | ÂÖÉÁ¥†Á∫ßÁöÑÈÄªËæë‰∏éÊìç‰Ωú               |
| `\|`    | Êåâ‰ΩçÊàñÔºàORÔºâ    | `(arr < 3) \| (arr > 8)`      | ÂÖÉÁ¥†Á∫ßÁöÑÈÄªËæëÊàñÊìç‰Ωú               |
| `~`     | Êåâ‰ΩçÈùûÔºàNOTÔºâ   | `~(arr > 5)`                  | ÂèçËΩ¨Â∏ÉÂ∞îÊé©Á†ÅÔºàmaskÔºâ             |
| `!=`    | ‰∏çÁ≠â‰∫é          | `arr != 0`                    | Á≠õÈÄâ‰∏çÁ≠â‰∫éÊüêÂÄºÁöÑÂÖÉÁ¥†             |

````python
#@title Exercises NumPy

arr_2d = np.arange(27).reshape(3, -1)
print(arr_2d)

# ‰ªªÂä° 1ÔºöÂØπÊØèÂàóËøõË°å‰∏≠ÂøÉÂåñÔºàÂáèÂéªÂàóÁöÑÂùáÂÄºÔºâ
col_means = arr_2d.mean(axis=0)                    # Ê≤ø axis=0ÔºàÂç≥ÂàóÔºâËÆ°ÁÆóÂùáÂÄº
centered_by_column = arr_2d - col_means            # ÂØπÊØè‰∏ÄÂàóÁöÑÊØè‰∏™ÂÖÉÁ¥†ÂáèÂéªËØ•ÂàóÁöÑÂùáÂÄº
print("\n‰ªªÂä°1Ôºö\n", centered_by_column)


# ‰ªªÂä° 2ÔºöÂ∞ÜÊØè‰∏ÄË°åÂΩí‰∏ÄÂåñÔºàÊÄªÂíå‰∏∫ 1Ôºâ
row_sums = arr_2d.sum(axis=1, keepdims=True)       # ËÆ°ÁÆóÊØè‰∏ÄË°åÁöÑÊÄªÂíåÔºå‰øùÊåÅÂàóÁªìÊûÑÔºàÁî®‰∫éÂπøÊí≠Ôºâ
normalized_by_row = arr_2d / row_sums              # ÊØèË°åÈô§‰ª•ÂÆÉËá™Â∑±ÁöÑÂíåÔºåÂÆûÁé∞ÂΩí‰∏ÄÂåñ
print("\n‰ªªÂä°2Ôºö\n", normalized_by_row)


# ‰ªªÂä° 3ÔºöÂ∞ÜÊØè‰∏™ÂÖÉÁ¥†ÊõøÊç¢‰∏∫ÂÆÉ‰∏éÊï∞ÁªÑÊï¥‰ΩìÂùáÂÄºÁöÑË∑ùÁ¶ª
overall_mean = arr_2d.mean()                       # ËÆ°ÁÆóÊï¥‰∏™Êï∞ÁªÑÁöÑÂùáÂÄº
distance_from_mean = np.abs(arr_2d - overall_mean) # ‰ΩøÁî® np.abs ËÆ°ÁÆóÊØè‰∏™ÂÖÉÁ¥†‰∏éÊï¥‰ΩìÂùáÂÄºÁöÑÁªùÂØπË∑ùÁ¶ª
print("\n‰ªªÂä°3Ôºö\n", distance_from_mean)
````
